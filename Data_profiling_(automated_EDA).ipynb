```python
# Core data libraries
import pandas as pd
import numpy as np

# automated EDA
from ydata_profiling import ProfileReport
```

```python
# Load datasets
viewership = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/viewership.csv")
user_profile = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/user_profile.csv")

# Preview data
viewership.head()
user_profile.head()
```

```python
print(viewership.info())
print(user_profile.info())
```

```python
# Example: joining on user_id
df = pd.merge(viewership, user_profile, on="UserID", how="left")
df = df.drop(columns=['Unnamed: 4', 'Unnamed: 5'])
df.head()
```

```python
profile = ProfileReport(df, explorative=True)
profile.to_notebook_iframe()
```
