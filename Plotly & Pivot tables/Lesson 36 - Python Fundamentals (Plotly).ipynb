```python
# Core data libraries
import pandas as pd
import numpy as np

# Visualization
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd

user_profile = spark.sql("select * from brightlearn.brighttv.user_profile").toPandas()
viewership = spark.sql("select * from brightlearn.brighttv.viewership").toPandas()

viewership.head()
user_profile.head()
```

```python
print(viewership.info())
print(user_profile.info())
```

# Join the Data

```python
df = pd.merge(viewership, user_profile, on="UserID", how="left")
df.head()
```

# Basic Aggregation

```python
df['duration'] = pd.to_timedelta(df['Duration 2'].dt.hour, unit='h') + \
                       pd.to_timedelta(df['Duration 2'].dt.minute, unit='m') + \
                       pd.to_timedelta(df['Duration 2'].dt.second, unit='s')


df['duration'] = pd.to_timedelta(df['duration'] )
df['session_duration'] = df['duration'].dt.total_seconds() / 60

df.head()
```

```python
# total watch time by gender
gender_watch = df.groupby("Gender")["session_duration"].sum().reset_index()

gender_watch
```

# Plotly

## Bar Chart

```python
fig = px.bar(
    gender_watch,
    x="Gender",
    y="session_duration",
    title="Total Watch Time by Gender",
    color="Gender"
)

fig.show()
```

## Time trend

```python
df["view_date"] = pd.to_datetime(df["RecordDate2"])
df.info()
```

```python
daily_trend = df.groupby("view_date")["session_duration"].sum().reset_index()
```

```python
fig = px.line(
    daily_trend,
    x="view_date",
    y="session_duration",
    title="DAily watch time trend"
)

fig.show()
```

```python
df["age_group"] = df["Age"].apply(lambda x: "Young" if x<= 17 else "Adult")
df.head()
```

```python
age_watch = df.groupby("age_group")["session_duration"].sum().reset_index()

fig = px.bar(
    age_watch,
    x="age_group",
    y="session_duration",
    title="Average watch time by age group",
    color = "age_group"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="Age",
    y="session_duration",
    title="Watch time vs Age",
    color = "Gender"
)

fig.show()
```
