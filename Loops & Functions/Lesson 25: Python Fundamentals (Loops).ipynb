# FOR LOOP

## For Loop (Iteration over a sequence)

A `for` loop goes through each item in a sequence one by one.

**Flow:**
- Start from first item
- Execute code
- Move to next item
- Stop when sequence ends


```python
numbers = [1,2,3,4,5]

for num in numbers:
  print(num)
```

```python
revenue = [100, 200, 50, 150, 400]

for r in revenue:
    print(r)
```

```python
revenue = [100, 200, 50, 150, 400]

for r in revenue:
    if r > 150:
        print(r)
```

### filtering without a loop

```python
list(map(print, filter(lambda r: r >150, revenue)))
```

# While loop

###  While Loop (Condition-based loop)

A `while` loop runs as long as a condition is TRUE.

**Flow:**
- Check condition
- If TRUE → run code
- Repeat
- If FALSE → stop

```python
counter = 0

while counter < 10:
  print("Hello word!")
  counter = counter + 1
```

```python
counter = 2

while counter < 10:
  print(counter, ": Hello word!")
  counter = counter + 2
```

```python
while (True):
    print("Bright Learn")
```

```python
revenue = [100, 200, 50, 150, 400, 100, 200]

total_revenue = 0 #Running total
i = 0 #counter

while total_revenue < 600:
    total_revenue = total_revenue + revenue[i]
    print(f"Added: {revenue[i]}, Running Total revenue = {total_revenue}")
    i = i + 1
```

# ⚖️ For Loop vs While Loop

| **For Loop** | **While Loop** |
|----------|------------|
| Iterates over a sequence | Runs based on condition |
| Known number of iterations | Unknown number of iterations |
| Safer (less risk of infinite loop) | Risk of infinite loop |
