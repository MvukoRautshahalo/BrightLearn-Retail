# Mini Exercise:

Create a function "discount" that will input the original amount and return a discounted amount:

def discount(amount):



Rules:
- <100 --> 0%
- 100 - 300 --> 10%
- 300 --> 20%



Results:
- dataframe with original columns
- add new columns to dataframe
-         - Discounted Amount
-         - Discount

```python
discounts = []
discounted_amounts = []

for i in range(len(df)):
    amount = df.loc[i, "Total Amount"]
    
    if amount < 100:
        discount = 0
    elif amount <= 300:
        discount = 0.1
    else:
        discount = 0.2
        
    discounted_revenue = amount * (1 - discount)
    
    discounts.append(discount)
    discounted_amounts.append(discounted_revenue)

df["Discount"] = discounts
df["Discounted Amount"] = discounted_amounts
```
