# zip return object iterable tuples

```python
for category,qty, price in zip(df["Product Category"],df["Quantity"],df["Price per Unit"]):
    revenue = qty* price
    print(category,revenue)
```

# enumerate - adds a counter 

```python
for i, category in enumerate(df["Product Category"]):
    print(f"{i}. {category}")
```

```python
for i, (category,qty) in enumerate(zip(df["Product Category"], df["Quantity"])):
    print(f"{i}. {category} - {qty} units")
```

# Nested loops

Loops inside a loop


Syntax:

`Outer_loop Expression:`

    Inner_loop Expression:
        Statement inside inner_loop
    Statement inside Outer_loop

```python
for cat1 in df["Product Category"]:
  for cat2 in df["Product Category"]:
    if cat1 != cat2:
      print(cat1, "is bought with", cat2)
  
```

```python
categories = df["Product Category"].unique().tolist()
genders = df["Gender"].unique().tolist()
```

```python
categories
```

```python
genders
```

Category (outer loop)  - Gender (inner loop)

```python
for category in categories:
    for gender in genders:
        condition = (df["Product Category"] == category) & (df["Gender"] == gender)
        revenue = df.loc[condition, "Total Amount"].sum()
        print(f" {category:<15} | {gender:<8} | R{revenue:>10,.0f}")
```

label transaction spend

```python
def classify_Spend(amt):
    if amt < 100: return "Low"
    elif amt < 500: return "Medium"
    else: return "High"
```

```python
spend_labels = [classify_Spend(amt) for amt in df["Total Amount"]]
df["Spend_labels"] = spend_labels

df.head()
```
