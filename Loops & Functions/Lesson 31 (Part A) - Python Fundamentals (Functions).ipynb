# Lambda function

- A lambda function in Python is a small, anonymous function defined without a name. 
- Unlike standard functions defined with the def keyword, lambda functions are typically used for short-lived, one-off operations where a full function definition would be unnecessary.

```python
def square(x):
    return x*x

square = square(x)
```

```python
sqaure = lambda x: x*x
```

- lambda -> creates a function
- x -> input
- x*x -> logic / output

```python
# 1. Define a list of integers to be processed
numbers = [1, 2, 3, 4, 5]

# 2. Use map() to apply a lambda function to every item in 'numbers'
# The lambda function (lambda x: x * x) squares each element
# list() converts the resulting map object back into a standard list
squared = list(map(lambda x: x * x, numbers))

# 3. Print the final list of squared numbers to the console
print(squared)
```

```python
numbers = [1,2,3,4,5]

list2 = [1,2,3,4,5]

sqaured = list(map(lambda x,y: x + y, numbers, list2))

print(sqaured)
```

```python
lst_spend = [100, 600, 1500, 400]

segments = list(map(lambda x:
    "High" if x > 1000 else "Medium" if x > 500 else "Low", lst_spend))

print(segments)
```
