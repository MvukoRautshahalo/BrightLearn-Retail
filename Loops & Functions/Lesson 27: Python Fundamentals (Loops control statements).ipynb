# Loop control statements

 
- Includes _break, continue, and pass,_ 
- They alter the normal flow of loops (for, while) by enabling early termination, skipping iterations, or acting as placeholders.

---

| **Statement** |	**Purpose	Action**|
| ----------|---------------|
| break     |	Exits loop completely	Stops the loop immediately.|
| continue	| Skips current iteration	Jumps to the next iteration. |
| pass	    | Acts as a placeholder	Does nothing, continues to next line.|

```python
# Count high value transactions

count = 0

for i in range(len(df)):
  if df.loc[i, "Total Amount"] > 300:
    count += 1
    count = count + 1

print("Transactions > 300: ", count)
```

> ## Stop Loop Early (break)

```python
Students = ['John', 'Jane', 'Mvuko', 'Dawn']

for name in Students:
  if name == 'Mvuko':
    break

  print(name)
```

```python
for i in range(len(df)):

    if df.loc[i, "Product Category"] == "Electronics":
        break

    print(df.loc[i, "Product Category"])
```

> ## Skip Value - [continue]

```python
Students = ['John', 'Jane', 'Mvuko', 'Dawn']

for name in Students:
  if name == 'Mvuko':
    continue

  print(name)
```

```python
for i in range(len(df)):
    amount = df.loc[i, "Total Amount"]

    if amount < 50:
        continue

    print(amount)
```
