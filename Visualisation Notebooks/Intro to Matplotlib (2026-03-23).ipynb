```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spark_df = spark.sql("select * from default.retail_sales_dataset")
df = spark_df.toPandas()

df.head()
```

```python
df.info()
```

```python
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()
print("REvenue by Product: ")
print(revenue_by_product)
```

#First Visualisation

```python
revenue_by_product.plot(kind = "bar")
plt.title("Revenue by Product")
plt.xlabel("Product Categories")
plt.ylabel("Revenue")
plt.xticks(rotation = 45)
```

#Time based analysis (Line chart)

```python
df["date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
revenue_over_time = df.groupby("date")["Total Amount"].sum()
```

```python
revenue_over_time.plot(kind ="line")
plt.title("Revenue Over TIme")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```
