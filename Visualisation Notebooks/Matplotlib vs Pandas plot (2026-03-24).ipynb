# Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import data

```python
spark_df = spark.sql("select * from default.retail_sales_dataset")
df = spark_df.toPandas()

df.head()
```

# Explore data

```python
df.info()
```

```python
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()
print("REvenue by Product: ")
print(revenue_by_product)
```

# Matplotlib with pandas

```python
revenue_by_product.plot(kind = "bar")
plt.title("Revenue by Product")
plt.xlabel("Product Categories")
plt.ylabel("Revenue")
plt.xticks(rotation = 45)
```

#Time based analysis (Line chart)

```python
df["date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
revenue_over_time = df.groupby("date")["Total Amount"].sum()
```

```python
revenue_over_time.plot(kind ="line")
plt.title("Revenue Over TIme")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

# Figure + Axes

- figure - (fg): canvas, page
- Axes - (ax): plot area, chart on the page

#Multi-Line plot

- unstack() - converts the product categories into columns
- Each column is a separate line

# Multi - plot (side-byside)

```python
revenue_over_time = df.groupby("date")["Total Amount"].sum()
```

```python
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()
```

```python
fig, ax = plt.subplots(1,2, figsize =(12,5))
```

```python
fig, ax = plt.subplots(1,2, figsize =(12,5))
#Left graph
ax[0].bar(revenue_by_product.index, revenue_by_product.values)

ax[0].set_title("Revenue by product")
ax[0].set_xlabel("Product")
ax[0].set_ylabel("Revenue")
ax[0].tick_params(axis = 'x', rotation=45)
ax[0].grid(True)

#Right graph
ax[1].plot(revenue_over_time.index, revenue_over_time.values)

ax[1].set_title("Revenue over time")
ax[1].set_xlabel("date")
ax[1].set_ylabel("Revenue")
ax[1].tick_params(axis = 'x')
ax[1].grid(True)

plt.tight_layout()

plt.show()
```
