```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

# EDA

```python
spark_df = spark.sql("select * from default.retail_sales_dataset")
df = spark_df.toPandas()

# Convert date column
df["date"] = pd.to_datetime(df["Date"])

df.head()
```

```python
df.info()
```

# Graphs

```python
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()
```

## Barplot

```python
revenue_by_product.plot(kind = "bar")
```

```python
sns.set_style("whitegrid")
```

```python
sns.barplot(x=revenue_by_product.index, y=revenue_by_product.values)
plt.title("Revenue by product")
plt.xticks(rotation = 45)
plt.show()
```

## Line plot

```python

revenue_over_time = df.groupby("date")["Total Amount"].sum().reset_index()

sns.lineplot(data = revenue_over_time, x="date", y = "Total Amount")

plt.show()
```

## Multi-line plot

```python
sns.lineplot(data=df, x="date", y="Total Amount", hue="Product Category")
```

```python
pivot = df.groupby(["date","Product Category"])["Total Amount"].sum().unstack()
pivot.plot()
plt.show()
```

## Pie chart

### Pie charts are risky. Only use them for proportions.

- values - sizes of slices
- labels - names
- autopct - shows %
- startangle - rotates chart

```python
plt.figure(figsize=(6,6))

plt.pie( revenue_by_product.values,
        labels = revenue_by_product.index,
        autopct='%1.1f%%',
        startangle = 90
        )

plt.title("Revenue contribution by Product")
plt.show()
```

## Primary & Secondary axes : Dual -Axis Chart

```python
revenue = df.groupby("date")["Total Amount"].sum()
transactions = df.groupby("date")["Quantity"].sum()
```

```python

# Create figure
fig, ax1 = plt.subplots(figsize=(10,5))

# -----------------------------
# First axis (Revenue)
# -----------------------------
ax1.plot(revenue.index, revenue.values, color="green")
ax1.set_xlabel("Date")
ax1.set_ylabel("Revenue")
ax1.tick_params(axis='y', labelcolor="green")

# -----------------------------
# Second axis (Transactions)
# -----------------------------
ax2 = ax1.twinx()

ax2.bar(transactions.index, transactions.values, alpha=0.3)
ax2.set_ylabel("Number of Transactions")

# Title
plt.title("Revenue vs Transactions Over Time")

plt.show()
```
