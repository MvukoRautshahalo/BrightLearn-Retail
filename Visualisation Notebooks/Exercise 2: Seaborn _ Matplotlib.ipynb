```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spark_df = spark.sql("select * from default.retail_sales_dataset")
df = spark_df.toPandas()

# Convert date column
df["date"] = pd.to_datetime(df["Date"])

df.head()
```

```python

# -----------------------------
# Create combined figure
# -----------------------------
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# -----------------------------
# LEFT: Boxplot
# -----------------------------
sns.boxplot(data=df, x="Product Category", y="Total Amount", ax=ax[0])

ax[0].set_title("Revenue Distribution by Product")
ax[0].tick_params(axis='x', rotation=45)

# -----------------------------
# RIGHT: Histogram
# -----------------------------
sns.histplot(data=df, x="Total Amount", hue="Product Category", bins=20, ax=ax[1], kde=True)

ax[1].set_title("Revenue Distribution (All Products)")

plt.savefig('my_plot.png', bbox_inches='tight', dpi=300)
plt.tight_layout()
plt.show()
```

# 1. Boxplot - Variability + Outliers

- Median (middle line)
- Spread (box = IQR)
- Outliers (points outside whiskers)

## Which product has highest variability?

```python
sns.boxplot(data=df, x="Product Category", y="Total Amount")
```

# 2. Histogram — Distribution Shape

What it shows:
- Frequency of revenue values
- Distribution shape
- Overlap between products

```python
sns.histplot(data=df, x="Total Amount", hue="Product Category", kde=True)
```

# Insights

**Insight 1 — Distribution Shape (Histogram)**

The revenue distribution is right-skewed, with the majority of transactions concentrated at lower values and a long tail extending toward higher values.

Evidence:
- High frequency in low-value bins
- KDE curve stretched to the right

Meaning:
- Most transactions are small
- Few high-value transactions drive the upper range

**Insight 2 — Variability (Boxplot)**

Electronics and Beauty exhibit higher variability compared to Clothing, as indicated by wider interquartile ranges and longer whiskers.

Evidence:
- Larger box (IQR) for Electronics and Beauty
- Longer whiskers

Meaning:
- Revenue values are more spread out
- Less consistency in transaction amounts

**Insight 3 — Outliers (Boxplot)**

Clothing contains distinct high-value outliers, visible as points beyond the upper whiskers, indicating occasional extreme transactions.

Evidence:
- Individual points above whiskers in Clothing
- Less frequent but significantly higher values

Meaning:
- Irregular spikes in revenue
- Potential anomalies or premium transactions
