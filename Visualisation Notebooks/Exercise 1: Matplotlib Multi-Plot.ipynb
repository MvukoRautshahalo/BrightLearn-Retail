```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spark_df = spark.sql("select * from default.retail_sales_dataset")
df = spark_df.toPandas()

df.head()
```

```python
df.info()
```

```python

# Convert date column
df["date"] = pd.to_datetime(df["Date"])

# Create figure
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# -----------------------------
# 1. Bar Chart (Revenue by Product)
# -----------------------------
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()

ax[0, 0].bar(revenue_by_product.index, revenue_by_product.values)
ax[0, 0].set_title("Revenue by Product")
ax[0, 0].set_xlabel("Product Category")
ax[0, 0].set_ylabel("Revenue")
ax[0, 0].tick_params(axis='x', rotation=45)

# -----------------------------
# 2. Line Chart (Revenue Over Time)
# -----------------------------
revenue_over_time = df.groupby("date")["Total Amount"].sum()

ax[0, 1].plot(revenue_over_time.index, revenue_over_time.values)
ax[0, 1].set_title("Revenue Over Time")
ax[0, 1].set_xlabel("date")
ax[0, 1].set_ylabel("Total Amount ")

# -----------------------------
# 3. Histogram (Distribution)
# -----------------------------
ax[1, 0].hist(df["Total Amount"], bins=20)
ax[1, 0].set_title("Revenue Distribution")
ax[1, 0].set_xlabel("Total Amount")

# -----------------------------
# 4. Multi-line Chart (Product Trends)
# -----------------------------
pivot = df.groupby(["date", "Product Category"])["Total Amount"].sum().unstack()

for col in pivot.columns:
    ax[1, 1].plot(pivot.index, pivot[col], label=col)

ax[1, 1].set_title("Revenue by Product Over Time")
ax[1, 1].legend()

# Layout fix
plt.tight_layout()
plt.show()
```
