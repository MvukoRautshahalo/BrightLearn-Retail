# Market Basket Analysis
We are working like analysts at Pick n Pay.

Goal: Find products bought together and measure relationships.

```python
import pandas as pd#for working with data
from itertools import combinations# To create pairs of items
from collections import Counter # to count how often each pair appears
import matplotlib.pyplot as plt #visualisation
```

## Load Data

```python
df = pd.read_csv('market_basket_dataset.csv')
df.head()
```

## Create Baskets

```python
# Group all products by customerID
# This creates a list of items each customer bought
basket = df.groupby('CustomerID')['Product'].apply(list)
basket
```

## Generate Pairs

```python
pairs = [] # create an empty list to store all pairs

# Loop through each customer's basket
for items in basket:
    pairs.extend(combinations(items, 2))

    # Generate all possible pairs of products (2 at a time)
    # Example: [Bread, Milk, Eggs] → (Bread, Milk), (Bread, Eggs), (Milk, Eggs)
```

```python
pairs
```

## Count Pairs

```python
pair_counts
```

```python
# Count how many times each pair appears
pair_counts = Counter(pairs)

# Convert to DataFrame for easier analysis
pairs_df = pd.DataFrame(pair_counts.items(), columns=['Pair', 'Count'])

# Sort from most frequent to least frequent
pairs_df = pairs_df.sort_values(by='Count', ascending=False)

pairs_df.head()
```

## Calculate Support
Support = Count / Total Transactions

```python
len(basket)
```

```python
# Total number of transactions (customers)
total = len(basket)

# Support = how often a pair appears out of all transactions
pairs_df['Support'] = pairs_df['Count'] / total

pairs_df.head()
```

## Calculate Confidence
Confidence(A → B) = (A & B) / A

Interpretation:
- If someone buys A, how likely are they to also buy B?

```python
basket
```

```python
#Example: buying Milk given already bought Bread
# Count how many baskets contain Bread
bread_count = sum(['Bread' in items for items in basket])

# Count how many baskets contain BOTH Bread and Milk
bread_milk = pair_counts.get(('Bread', 'Milk'), 0)

# Confidence = probability of buying Milk given Bread
confidence = bread_milk / bread_count

confidence
```

## Calculate Lift
Lift = Confidence / Support(B)

Interpretation:
- Lift > 1 → strong relationship
- Lift = 1 → no relationship
- Lift < 1 → weak relationship

```python
# Count how many baskets contain Milk
milk_count = sum(['Milk' in items for items in basket])

# Support of Milk alone
support_milk = milk_count / len(basket)

# Lift = strength of relationship
lift = confidence / support_milk

lift
```

## Visualize Top Pairs

```python
# Plot top product pairs
pairs_df.head(5).plot(x='Pair', y='Count', kind='bar')

plt.title("Top Product Pairs")
plt.xlabel("Product Pair")
plt.ylabel("Frequency")

plt.show()
```
